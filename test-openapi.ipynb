```python
from openai import OpenAI

# The client automatically looks for the 'OPENAI_API_KEY' environment variable
client = OpenAI()

try:
    response = client.chat.completions.create(
        model="gpt-5-mini",  # Using the stable mini model for a quick test
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": "Confirm if you can hear me!"}
        ]
    )
    print("Success! Response from AI:")
    print(response.choices[0].message.content)

except Exception as e:
    print(f"Connection failed: {e}")
```
